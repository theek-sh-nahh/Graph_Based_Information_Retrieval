```python
import os
import random

DATA_FOLDER = "data"
os.makedirs(DATA_FOLDER, exist_ok=True)

topics = {
    "information retrieval": [
        "Information retrieval systems rank documents using relevance scoring algorithms.",
        "Search engines rely on indexing and retrieval models to process large document collections.",
        "TF-IDF and BM25 are classical ranking methods in retrieval systems.",
        "Query expansion techniques improve retrieval effectiveness.",
        "Relevance feedback enhances document ranking accuracy.",
        "Inverted indexes allow efficient term-based searching."
    ],
    "machine learning": [
        "Machine learning models learn patterns from data for predictive tasks.",
        "Supervised learning requires labeled training data.",
        "Neural networks are widely used in classification problems.",
        "Support vector machines maximize margin between classes.",
        "Feature engineering improves model performance.",
        "Gradient descent is used to optimize model parameters."
    ],
    "data analytics": [
        "Data analytics involves extracting insights from structured and unstructured data.",
        "Exploratory data analysis helps identify trends and anomalies.",
        "Visualization techniques support better decision making.",
        "Big data systems process high-volume datasets efficiently.",
        "Statistical analysis supports hypothesis testing.",
        "Predictive analytics forecasts future trends."
    ]
}

def generate_document():
    selected_topics = random.sample(list(topics.keys()), k=2)
    sentences = []

    for topic in selected_topics:
        sentences.append(random.choice(topics[topic]))
        sentences.append(random.choice(topics[topic]))

    random.shuffle(sentences)
    return " ".join(sentences)


# Count existing docs
existing_docs = [f for f in os.listdir(DATA_FOLDER) if f.endswith(".txt")]
start_index = len(existing_docs) + 1

# Generate until 100 total
target_total = 100
docs_to_generate = target_total - len(existing_docs)

for i in range(docs_to_generate):
    content = generate_document()
    filename = f"doc{start_index + i}.txt"
    
    with open(os.path.join(DATA_FOLDER, filename), "w", encoding="utf-8") as f:
        f.write(content)

print(f"{docs_to_generate} documents generated successfully.")
```
